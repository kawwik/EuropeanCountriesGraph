```python
# (d) minimum edge coloring
edgeColoring = graph_coloring.edge_coloring(graph) # находим раскраску
colors = rainbow(len(edgeColoring))
edge_colors = {}
for i in range(len(edgeColoring)): 
    edge_colors[colors[i]] = edgeColoring[i]
    
pl = graph.plot(edge_colors = edge_colors)
pl.show()
```

```python
# (f) maximum stable set
stableSet = graph.independent_set() # находим независимое множество вершин
stableSetColoring = {'#FF4500' : stableSet}
pl = graph.plot(vertex_colors = stableSetColoring)
pl.show()
```

```python
# (g) maximum matching
edgeMatching = graph.matching() # находим независимое множество рёбер
matchingColors = {'#FF4500' : edgeMatching}
pl = graph.plot(edge_colors = matchingColors, edge_thickness = 2)
pl.show()
```

```python
# (h) minimum vertex cover
vertexCover = graph.vertex_cover() # находим наименьшее вершинное покрытие
vertexCoverColoring = {'#FF4500' : vertexCover}
pl = graph.plot(vertex_colors = vertexCoverColoring)
pl.show()
len(vertexCover)
```

```python
# (l) blocks and block-cut tree
blocks = graph.blocks_and_cut_vertices()[0] # находим блоки графа
colors = rainbow(len(blocks))
blocksColoring = {}
for i in range(len(blocks)): # раскрашиваем блоки и визуализиурем
    blocksColoring[colors[i]] = blocks[i]
    
pl = graph.plot(vertex_colors = blocksColoring)
pl.show()

# создаём block-cut tree и переименовываем вершины, приводя граф к читабельному виду
blockCutsTree = Graph(graph.blocks_and_cuts_tree())
blockCutsTree.plot().show()
blockCutsTree.relabel(['A','B','C','D','E','F','G','H', 12, 13, 16, 31, 36])
blockCutsTree.plot().show()
```

```python
# Рёберная двусвязность
graphWithNoBridges = graph.copy()
graphWithNoBridges.delete_edges(graph.bridges()) # находим и удаляем мосты
comps = graphWithNoBridges.connected_components() # находим компоненты связности в полученном графе
colors = rainbow(len(comps))
compColors = {}
for i in range(len(comps)): # раскрашиваем и визуализируем их
    compColors[colors[i]] = comps[i]
graph.plot(vertex_colors = compColors).show()
```
